```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, f_oneway, chi2_contingency,binom_test,pearsonr
```

#### Data reading and inspection section

```python
df = pd.read_csv('sampling 1 plays.csv', error_bad_lines = False, low_memory=False)
print(len(df))
df.drop_duplicates(subset='hash_watcher_id',inplace=True)
print(len(df))
df['os_name'].fillna('Unknown',inplace=True)
df['play_time']=df['play_time'].astype('datetime64[ns]')
df['days']=df.play_time.dt.day
display(df.head())
```

```python
perday = df.groupby(['days','hash_watcher_id'],as_index=False).play_duration.sum().reset_index(drop=True)
best = perday.groupby('days').apply(\
        lambda x: pd.Series({
    'player':x['hash_watcher_id'],
    'duration': x['play_duration'].max()
}))

display(best)
```

```python
d1 = perday.loc[(perday.days == 1)].reset_index(drop=True)
d1.sort_values(by='play_duration',ascending=False,inplace=True)
best_day1 = d1.iloc[0:10]

display(best_day1)
```

```python
days = df.play_time.dt.day.unique()
days.sort()
def best_10(day):
    if day in days:
        dx = perday.loc[(perday.days == day)].reset_index(drop=True)
        dx.sort_values(by='play_duration',ascending=False,inplace=True)
        best_dayx = dx.iloc[:10]
        return best_dayx
    else:
        return 'please input correct day available'

for i in days:
    display(best_10(i))
```

```python
#play duration on each platform
windows_play = df.loc[(df.os_name == 'Windows')]['play_duration'].reset_index(drop=True)
windows_play = pd.DataFrame(windows_play)
wp = windows_play[~(windows_play == 0).any(axis=1)].reset_index(drop=True)
ios_play = df.loc[(df.os_name == 'iOS')]['play_duration'].reset_index(drop=True)
ios_play = pd.DataFrame(ios_play)
ip = ios_play[~(ios_play ==0).any(axis=1)].reset_index(drop=True)
android_play = df.loc[(df.os_name == 'Android')]['play_duration'].reset_index(drop=True)
android_play = pd.DataFrame(android_play)
ap = android_play[~(android_play ==0).any(axis=1)].reset_index(drop=True)
win_avg = np.mean(wp.play_duration.values)
win_std = np.std(wp.play_duration.values)
```

```python
q3 = np.quantile(wp,0.75)
q1 = np.quantile(wp,0.25)
iqr = q3-q1
print(iqr)

fig = plt.figure(figsize=(12,14))
ax = plt.subplot2grid((6,6),(0,0),colspan=3,rowspan=3)
plt.hist(wp.play_duration.values,bins=20,label='Windows user \n playtime')
plt.legend()
ax1 = plt.subplot2grid((6,6),(0,3),colspan=3,rowspan=3)
plt.hist(ip.play_duration.values,bins=20,color='green',label='iOS user \n playtime')
plt.legend()
ax2 = plt.subplot2grid((6,6),(3,0),rowspan=3,colspan=6)
plt.hist(ap.play_duration.values,bins=20,color='purple',label='Android user \n playtime')
fig.tight_layout(pad=1)
plt.legend()
plt.show()
```

```python
platform_and_app = df.groupby(['browser_name']).apply \
    (lambda x: pd.Series({
        'play_counts': x['play_duration'].sum(),
        'user_counts': x['player_name'].count()
        })).reset_index()

platform_and_app_sorted = platform_and_app.sort_values(by=['play_counts','user_counts'],ascending=False).reset_index(drop=True)
display(platform_and_app_sorted)
```

```python
platform_and_users = df.groupby(['os_name']).apply \
    (lambda x: pd.Series({
    'play_counts': x['play_duration'].sum(),
    'user_counts': x['player_name'].count()
})).reset_index()

sort_PU=platform_and_users.sort_values(by=['play_counts','user_counts'],ascending=False).reset_index(drop=True)
sort_PU['percentage'] = sort_PU.apply(lambda x: x['user_counts'] / len(df) *100.0,axis=1)
sort_PU.loc[sort_PU.percentage <=3,'os_name'] = 'Others'
sort_PU.loc[sort_PU.percentage <=3, 'play_counts'] = sort_PU.play_counts[4:].sum()
sort_PU.loc[sort_PU.percentage <=3, 'user_counts'] = sort_PU.user_counts[4:].sum()
sort_PU.loc[sort_PU.percentage <=3, 'percentage'] = sort_PU.percentage[4:].sum()
sort_PU.drop_duplicates(subset='os_name',inplace=True)
display(sort_PU)
```

```python
label = sort_PU.os_name.values
plt.figure(figsize=(12,12))
ax= plt.subplot()
plt.pie(sort_PU.user_counts.values,labels=label,autopct='%d%%')
plt.title('Platform users')
plt.legend(label)
plt.show()
```

### Is there any correlationship between amount of users and play duration within os platform?

```python
list1= sort_PU.play_counts
list2= sort_PU.user_counts

#corr test
r,p = pearsonr(list1,list2)
print(r)
print(p)
```
